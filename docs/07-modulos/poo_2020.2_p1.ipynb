### Classe `Professor`

Defina a classe `Professor` para representar um professor que atua em uma universidade.

Atributos:

- `nome`, para representar o nome do professor (encapsulado com get/set).
- `departamento`, para representar o nome do seu departamento (encapsulado com get/set).

Métodos:

Com exceção dos get/set solicitados a classe não possui nenhum método.

### Classe `Disciplina`

Defina a classe `Disciplina` para representar uma disciplina em uma universidade.

Atributos:

- `nome`, para representar o nome da disciplina (privado).
- `codigo`, para representar o código da disciplina (privado).

Métodos:

- `__init__`: deve receber como parâmetros o seu nome e código
- `__str__`: converte a disciplina para uma string, contendo o seu código, nome e o professor responsável (string `'N/A'` caso não exista)
- `adiciona_prerequisito`: adiciona uma `Disciplina` como o seu pré-requisito
- `tem_prerequisito`: retorna verdadeiro se o objeto tem como pré-requisito uma outra `Disciplina` passada como parâmetro e falso caso contrário
- `associa_professor`: associa um `Professor` passado como parâmetro ao objeto
- O professor responsável pela disciplina deve ser acessado via método get

Utilize o código base a seguir como ponto de partida.

### Classe `Curso`

Defina a classe `Curso` para representar um curso oferecido por uma universidade.

Atributos:

- `nome`, para representar o nome do curso (privado).

Métodos:

- `__init__`: deve receber como parâmetros o nome e uma lista de objetos da classe `Disciplina` que fazem parte do curso
- `__str__`: converte o curso para uma string, contendo o seu nome e todas as suas disciplinas
- `remove_disciplina`: remove do curso a `Disciplina` que tenha como nome a string passada como parâmetro
- `busca_disciplinas`: retorna uma lista com todas as listas ministradas por um professor passado como parâmetro. Para realizar a busca, acesse o atributo encapsulado do tipo `Professor` da classe `Disciplina`. Observe também o caso da disciplina não possuir um `Professor` associado a ela.
